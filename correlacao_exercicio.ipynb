# Correlação

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dados de exemplo para correlação positiva
profundidade = np.linspace(10, 50, 100)
temperatura = 2 * profundidade + 5 + np.random.normal(0, 5, 100)
```

```python
temperatura
```

### Calcule o coeficiente de correlação entre a profundidade e a temperatura. Além disso plote Profundidade x Temperatura

```python
def cor(listax, listay):

    mediax = sum(listax)/len(listax)
    mediay = sum(listay)/len(listay)

    desviox = []
    for i in listax:
        dx = i - mediax
        desviox.append(dx)
    
    desvioy = []
    for i in listay:
        dy = i - mediay
        desvioy.append(dy)

    desvio2x = (np.array(desviox))**2
    desvio2y = (np.array(desvioy))**2

    desvioxy = np.array(desviox) * np.array(desvioy)

    r = sum(desvioxy)/(sum(desvio2x)*sum(desvio2y))**(1/2)

    return round(r,2)

    
```

```python
cor(profundidade, temperatura)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(profundidade, temperatura, color='blue', alpha=0.7)
plt.title('Correlação Positiva: Profundidade vs. Temperatura')
plt.xlabel('Profundidade (m)')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.show()

# Calcular e imprimir o coeficiente de correlação
corr_pos = np.corrcoef(profundidade, temperatura)[0, 1]
print(f'Coeficiente de correlação: r = {corr_pos:.2f}')
```

### Faça o mesmo para os dois conjuntos de dados 

```python
# Dados de exemplo para correlação negativa
teor_umidade = np.linspace(5, 30, 100)
resistividade = -3 * teor_umidade + 150 + np.random.normal(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.scatter(teor_umidade, resistividade, color='red', alpha=0.7)
plt.title('Correlação Negativa: Teor de Umidade vs. Resistividade')
plt.xlabel('Teor de Umidade (%)')
plt.ylabel('Resistividade (ohm-m)')
plt.grid(True)
plt.show()

# Calcular e imprimir o coeficiente de correlação
corr_neg = np.corrcoef(teor_umidade, resistividade)[0, 1]
print(f'Coeficiente de correlação: r = {corr_neg:.2f}')
```

```python
cor(teor_umidade, resistividade)
```

```python
# Dados de exemplo para ausência de correlação
sismos = np.random.uniform(10, 50, 100)
preco_petroleo = np.random.uniform(60, 90, 100)

plt.figure(figsize=(8, 6))
plt.scatter(sismos, preco_petroleo, color='green', alpha=0.7)
plt.title('Ausência de Correlação: Sismos vs. Preço do Petróleo')
plt.xlabel('Número de Sismos')
plt.ylabel('Preço do Petróleo (US$)')
plt.grid(True)
plt.show()

# Calcular e imprimir o coeficiente de correlação
corr_nula = np.corrcoef(sismos, preco_petroleo)[0, 1]
print(f'Coeficiente de correlação: r = {corr_nula:.2f}')
```

```python
cor(sismos,preco_petroleo)
```

```python
# Criar um DataFrame a partir dos dados gerados
data = {
    'Profundidade': profundidade,
    'Temperatura': temperatura,
    'Teor_Umidade': teor_umidade,
    'Resistividade': resistividade,
    'Sismos': sismos,
    'Preco_Petroleo': preco_petroleo
}
df = pd.DataFrame(data)

print("DataFrame com os dados:")
print(df.head())
```

```python
df
```

```python
# Correlação entre Profundidade e Temperatura
corr_prof_temp = df['Profundidade'].corr(df['Temperatura'])
print(f'\nCorrelação entre Profundidade e Temperatura: r = {corr_prof_temp:.2f}')

# Correlação entre Teor de Umidade e Resistividade
corr_umid_res = df['Teor_Umidade'].corr(df['Resistividade'])
print(f'Correlação entre Teor de Umidade e Resistividade: r = {corr_umid_res:.2f}')
```

```python
# Matriz de Correlação para o DataFrame completo
matriz_correlacao = df.corr()
print('\nMatriz de Correlação:')
print(matriz_correlacao.round(2))
```
